## Visualization

```python
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title("Distribution of Tenure")
plt.show()
```

```python
sns.countplot(x='Contract', data=df)
plt.title("Contract Type Distribution")
plt.xticks(rotation=45)
plt.show()
```

**
##Key Insights:
- Tenure shows a wide distribution, suggesting varying customer lifecycles.
- Month-to-month contracts dominate the dataset, which may correlate with higher churn risk.**
