# Data Preprocessing – Customer Churn

```python
#import libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df['customerID'].unique()
```

```python
df.info()
```

```python
df.head()
```

```python
# Column TotalCharges if numeric but its as object.
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.isna().sum()
```

```python
df['TotalCharges'].dtype
```

### Understanding 11 row:

```python
df_nan = df[df['TotalCharges'].isnull()]
```

```python
df_nan
```

```python
df[df['TotalCharges'].isna()][['tenure', 'MonthlyCharges', 'TotalCharges']]
```

**Observation since the tenure period is 0, they are new customer.**

```python
#Replacing NaN with 0:
df['TotalCharges'] = df['TotalCharges'].fillna(0)
df.isna().sum()
```

### Encoding ordinal, nominal and Target variable
Contract is Ordinal.
Other object type column are nominal.
Churn is Target variable


```python
#Binary Encoding
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df['Churn'].unique()
```

```python
#Ordinal Encoding
contract_map = {
    'Month-to-month': 0,
    'One year': 1,
    'Two year': 2
}

df['Contract'] = df['Contract'].map(contract_map)
```

```python
df['Contract'].unique()
```

```python
#Removing customerID column: It has 7023 unique value, while one hot encoding the column will be high.
#So, removing unwanted column.
df = df.drop(columns=['customerID'])
```

```python
#One hot endcoding
df = pd.get_dummies(df, drop_first=True)
```

```python
df.shape
```

```python
df.head()
```

## Saving the preprocessed data
use it for model building

```python
df.to_csv('../data/churn_cleaned.csv', index=False)
```
